```python
import fastf1

# Önbelleği (cache) aktif ediyoruz
fastf1.Cache.enable_cache('cache_verileri')

# 2023 Hollanda GP yarış verisini çekiyoruz
session = fastf1.get_session(2023, 'Zandvoort', 'R')
session.load()

# Max Verstappen'in en hızlı turunu buluyoruz
ver_lap = session.laps.pick_driver('VER').pick_fastest()

print(f"Max Verstappen'in kendi evindeki en hızlı tur zamanı: {ver_lap['LapTime']}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Uyarıyı dikkate alarak komutumuzu güncelledik
ver_lap = session.laps.pick_drivers('VER').pick_fastest()

# Bu turun telemetri (araç sensör) verilerini çekiyoruz
telemetry = ver_lap.get_telemetry()

# Pandas DataFrame ile veriyi tablo formatına getiriyoruz
# Şimdilik sadece Zaman, Hız, Vites ve Devir (RPM) kolonlarını görelim
df = pd.DataFrame(telemetry[['Time', 'Speed', 'nGear', 'RPM']])

# Tablonun ilk 5 satırını ekrana yazdıralım
print("--- Telemetri Verisi (İlk 5 Satır) ---")
print(df.head())
print("\nGrafik çiziliyor...")

# Basit bir Hız-Mesafe grafiği çizdirelim
plt.figure(figsize=(10, 5))
plt.plot(telemetry['Distance'], telemetry['Speed'], color='orange', linewidth=2)
plt.title("Max Verstappen - Zandvoort En Hızlı Tur Hız Profili")
plt.xlabel("Mesafe (metre)")
plt.ylabel("Hız (km/s)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import fastf1.plotting

# FastF1'in karanlık temasını ve takım renklerini aktif edelim
fastf1.plotting.setup_mpl()

# Fernando Alonso'nun en hızlı turunu çekiyoruz
alo_lap = session.laps.pick_drivers('ALO').pick_fastest()

# Alonso'nun telemetri verisi
alo_tel = alo_lap.get_telemetry()

# Red Bull ve Aston Martin'in resmi takım renklerini kütüphaneden çekiyoruz
rbr_color = fastf1.plotting.team_color('RBR')
ast_color = fastf1.plotting.team_color('AMR')

# Grafiği çizdirelim
plt.figure(figsize=(14, 6))

# Max'in çizgisi
plt.plot(telemetry['Distance'], telemetry['Speed'], color=rbr_color, label='VER - Red Bull', linewidth=2)

# Alonso'nun çizgisi
plt.plot(alo_tel['Distance'], alo_tel['Speed'], color=ast_color, label='ALO - Aston Martin', linewidth=2)

plt.title("Verstappen vs Alonso - Zandvoort 2023 Yarışı En Hızlı Tur Karşılaştırması")
plt.xlabel("Pist Üzerindeki Mesafe (metre)")
plt.ylabel("Hız (km/s)")
plt.legend()
plt.show()
```

```python
import fastf1
import fastf1.plotting
import matplotlib.pyplot as plt

# Önbelleği açık tutmaya devam ediyoruz
fastf1.Cache.enable_cache('cache_verileri')

# 1. RAPOR GÜNCELLEMESİ: Görselleştirme altyapısını ve temayı yeni standartla kuruyoruz
fastf1.plotting.setup_mpl(mpl_timedelta_support=True, color_scheme='fastf1')

# Oturumu seçiyoruz
session = fastf1.get_session(2023, 'Zandvoort', 'R')

# 2. RAPOR GÜNCELLEMESİ: Performans için hava durumu ve mesajları indirmekten vazgeçiyoruz
session.load(weather=False, messages=False)

# Turları alıyoruz
ver_lap = session.laps.pick_drivers('VER').pick_fastest()
alo_lap = session.laps.pick_drivers('ALO').pick_fastest()

# 3. RAPOR GÜNCELLEMESİ: Harita çizmeyeceğimiz için get_car_data ile yüksek performanslı telemetri çekiyoruz
# Ve add_distance() ile mesafe eksenini hesaplatıyoruz
ver_tel = ver_lap.get_car_data().add_distance()
alo_tel = alo_lap.get_car_data().add_distance()

# 4. RAPOR GÜNCELLEMESİ: Renkleri yeni dinamik mimariyle API'den çekiyoruz
ver_color = fastf1.plotting.get_team_color(ver_lap['Team'], session=session)
alo_color = fastf1.plotting.get_team_color(alo_lap['Team'], session=session)

# 4 Alt alta grafik oluşturuyoruz: Hız, Gaz, Fren, Vites
fig, ax = plt.subplots(4, 1, figsize=(16, 12), gridspec_kw={'height_ratios': [3, 1, 1, 1]})

# --- 1. HIZ (Speed) ---
ax[0].plot(ver_tel['Distance'], ver_tel['Speed'], color=ver_color, label='VER (Red Bull)', linewidth=2)
ax[0].plot(alo_tel['Distance'], alo_tel['Speed'], color=alo_color, label='ALO (Aston Martin)', linewidth=2)
ax[0].set_title("Zandvoort 2023: Verstappen vs Alonso - Tam Kapsamlı Telemetri Analizi")
ax[0].set_ylabel("Hız (km/s)")
ax[0].legend(loc='upper right')

# --- 2. GAZ (Throttle) ---
ax[1].plot(ver_tel['Distance'], ver_tel['Throttle'], color=ver_color, linewidth=2)
ax[1].plot(alo_tel['Distance'], alo_tel['Throttle'], color=alo_color, linewidth=2)
ax[1].set_ylabel("Gaz (%)")

# --- 3. FREN (Brake) ---
# Fren verisi basınca göre değişebilir, 0 (basılmıyor) ile 100 (tam fren) arasıdır
ax[2].plot(ver_tel['Distance'], ver_tel['Brake'], color=ver_color, linewidth=2)
ax[2].plot(alo_tel['Distance'], alo_tel['Brake'], color=alo_color, linewidth=2)
ax[2].set_ylabel("Fren (%)")

# --- 4. VİTES (nGear) ---
ax[3].plot(ver_tel['Distance'], ver_tel['nGear'], color=ver_color, linewidth=2)
ax[3].plot(alo_tel['Distance'], alo_tel['nGear'], color=alo_color, linewidth=2)
ax[3].set_ylabel("Vites")
ax[3].set_xlabel("Pist Üzerindeki Mesafe (metre)")

# Tüm grafiklere şık bir ızgara ekliyoruz
for a in ax:
    a.grid(True, linestyle='--', alpha=0.4)

# Grafikleri hizalayıp ekrana basıyoruz
plt.tight_layout()
plt.show()
```
